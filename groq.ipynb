# GROQ-API


```python
from langchain_groq import ChatGroq
from langchain_core.prompts import PromptTemplate
from langchain_core.output_parsers import StrOutputParser
from dotenv import load_dotenv
```

```python
load_dotenv()
```

```python
prompt = PromptTemplate(template="Translate the following text as Hello! How are you! from English to {mentioned} language",
                        input_variables=['mentioned'])
```

```python
llm = ChatGroq(model='llama-3.1-8b-instant')
```

```python
parser = StrOutputParser()
```

```python
chain = prompt | llm | parser
```

```python
translation=chain.invoke({'mentioned':'Hindi'})
```

```python
print(translation)
```
